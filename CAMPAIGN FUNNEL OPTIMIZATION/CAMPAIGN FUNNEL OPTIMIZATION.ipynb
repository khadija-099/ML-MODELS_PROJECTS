# DATA PREPARATION

```python
data= data.drop(['ID'],axis = 1)
data.head()
```

```python
data.info()
```

```python
data['DemAffl'] = data['DemAffl'].fillna(data['DemAffl'].mode()[0])
data['DemAge'] = data['DemAge'].fillna(data['DemAge'].mode()[0])
data['DemClusterGroup'] = data['DemClusterGroup'].fillna(data['DemClusterGroup'].mode()[0])
data['DemGender'] = data['DemGender'].fillna(data['DemGender'].mode()[0])
data['DemReg'] = data['DemReg'].fillna(data['DemReg'].mode()[0])
data['DemTVReg'] = data['DemTVReg'].fillna(data['DemTVReg'].mode()[0])
data['LoyalTime'] = data['LoyalTime'].fillna(data['LoyalTime'].mean())
```

# CONVERTING CATEGORY TO NUMERIC

```python
#labelEncoder
from sklearn.preprocessing import LabelEncoder

num = LabelEncoder()

data.DemClusterGroup = num.fit_transform(data.DemClusterGroup.astype('str'))
integer_mapping = {l: i for i, l in enumerate(num.classes_)}
print(integer_mapping)


data.DemGender = num.fit_transform(data.DemGender.astype('str'))
integer_mapping = {l: i for i, l in enumerate(num.classes_)}
print(integer_mapping)


data.DemReg = num.fit_transform(data.DemReg.astype('str'))
integer_mapping = {l: i for i, l in enumerate(num.classes_)}
print(integer_mapping)


data.DemTVReg = num.fit_transform(data.DemTVReg.astype('str'))
integer_mapping = {l: i for i, l in enumerate(num.classes_)}
print(integer_mapping)

data.LoyalClass = num.fit_transform(data.LoyalClass.astype('str'))
integer_mapping = {l: i for i, l in enumerate(num.classes_)}
print(integer_mapping)
```

# CHECKING FOR MULTICOLLINEARITY

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

def cal_vif(z):
    vif = pd.DataFrame()
    vif['variables'] = z.columns
    vif['VIF'] = [variance_inflation_factor(z.values, i) for i in range(z.shape[1])]
    
    return vif

```

```python
z = data.iloc[:,0:9]
cal_vif(z)
```

# VARIABLE SELECTION

```python
y = data.iloc[:, 9].values
x = data.iloc[:, 0:9].values
```

```python
x
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 0)
```

# MODELLING

```python
classifier = LogisticRegression(max_iter = 200)

classifier.fit(x_train,y_train)
y_pred = classifier.predict(x_test)
```

```python
y_pred
```

```python
import joblib
joblib.dump(classifier,'Classifier_LoyalCustomers')
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
predictions = classifier.predict_proba(x_test)
predictions
```

```python
#writing model output file
df_prediction_prob = pd.DataFrame(predictions,columns = ['prob_0','prob_1'])
df_test_data = pd.DataFrame(y_test,columns=['Actual Outcome'])
df_x_test = pd.DataFrame(x_test)

dfx = pd.concat([df_prediction_prob ,df_test_data, df_x_test], axis = 1)
```
